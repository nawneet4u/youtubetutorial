## Numerical Python(numpy) :
It is python library used to work with arrays. It provides various functions to create and work with arrays.

Difference between numpy array and list:
1.Numpy array stores values of same datatype,whereas list can store heterogenous values.
2.Numpy data structures takes less memory.
3.These are faster than lists.

```python
#It creates an array with evenly spaced elements.(startvalue,stopvalue(not included),step) 
x = np.arange(10,20,2) 
print(x)
```

```python
#It creates an array with evenly spaced elements with third value being number of elements to be generated.(startvalue,stopvalue,No of Elements)
c=np.linspace(1,100,11)
c
```

```python
# Creating array by a list
d=np.array([1,2,3,4,5])
d
```

```python
# Creating array by list name
e_list=[6,7,8,9,10]
e=np.array(e_list)
e
```

```python
#Creating 2D array by list
f=np.array([[1,2,3,4],[5,6,7,8]])
f
```

```python
#Creating 2D array by list name
g_list=[[1,2,3,4],[5,6,7,8]]
g=np.array(g_list)
g
```

```python
print(np.std(array_a))
```
